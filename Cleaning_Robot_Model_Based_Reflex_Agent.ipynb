State: 
0 = Sạch, 
1 = Bụi, 
2 = Robot, 
3 = Chướng ngại vật

```python
import random
```

```python
def rules(matrix, state):
    x = state[0]
    y = state[1]
    move = []

    if x > 0 and matrix[x-1][y] != 3:
        move.append("UP")

    if x < 2 and matrix[x+1][y] != 3:
        move.append("DOWN")

    if y > 0 and matrix[x][y-1] != 3:
        move.append("LEFT")

    if y < 2 and matrix[x][y+1] != 3:
        move.append("RIGHT")

    if len(move) == 0:
        return "STAY"

    return move
```

```python
def interpret_input(percept):
    state = []

    for i in range(3):
        for j in range(3):
            if percept[i][j] == 2:
                state = [i, j]

    return state
```

```python
def update_state(state, action, percept, model):
    new_state = interpret_input(percept)

    model["forbidden"] = None

    if action == "UP":
        model["forbidden"] = "DOWN"

    elif action == "DOWN":
        model["forbidden"] = "UP"

    elif action == "LEFT":
        model["forbidden"] = "RIGHT"

    elif action == "RIGHT":
        model["forbidden"] = "LEFT"

    return new_state
```

```python
def rule_match(matrix, state, model):
    move = rules(matrix, state)
    
    if move == "STAY":
        return "STAY"
    
    forbidden = model["forbidden"]

    if forbidden in move:
        move.remove(forbidden)
        
    if len(move) == 0:
        return "STAY"
    
    return random.choice(move)
```

```python
def move_robot(matrix, state, action):
    if action == "STAY":
        return
    
    x = state[0]
    y = state[1]

    matrix[x][y] = 0

    if action == 'UP':
        x -= 1
    elif action == 'DOWN':
        x += 1
    elif action == 'LEFT':
        y -= 1
    elif action == 'RIGHT':
        y += 1

    matrix[x][y] = 2
```

```python
def operation(matrix, state, action, model):
    percept = matrix
    state = update_state(state, action, percept, model)
    action = rule_match(matrix, state, model)
    move_robot(matrix, state, action)
    return state, action
```

```python
matrix = [[1, 1, 1],
        [1, 1, 1],
        [1, 1, 1]]

# Vị trí máy hút bụi ngẫu nhiên
matrix[random.randint(0,2)][random.randint(0,2)] = 2

# Vị trí chướng ngại vật ngẫu nhiên (2 cái)
for _ in range(2):
    if matrix[random.randint(0,2)][random.randint(0,2)] == 1:
        matrix[random.randint(0,2)][random.randint(0,2)] = 3
        
state = interpret_input(matrix)

action = None

model = {"forbidden": None}

step = 0
while True:
    print("Step: ", step)
    print("Trạng thái:")
    for row in matrix:
        print(row)

    state, action = operation(matrix, state, action, model)

    print("-" * 30)
    print("Hành động:", action)
    print("\n")

    count = 0
    for row in matrix:
        count += row.count(1)

    if count == 0:
        print("Đã hút sạch bụi!")
        for row in matrix:
            print(row)
        break
    
    step += 1
```
