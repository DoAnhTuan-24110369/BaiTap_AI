```python
import random
from collections import deque
```

```python
class Node:

    def __init__(self, state, parent=None, action=None):
        self.state = state
        self.parent = parent
        self.action = action
```

```python
GOAL = [
    [1,2,3],
    [4,5,6],
    [7,8,0]
]
```

```python
def goal_test(state):
    return state == GOAL
```

```python
def find_blank(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                return [i, j]
```

```python
def actions(state):
    x, y = find_blank(state)

    move = []

    if x > 0:
        move.append("UP")
    if x < 2:
        move.append("DOWN")
    if y > 0:
        move.append("LEFT")
    if y < 2:
        move.append("RIGHT")

    return move
```

```python
def result(state, action):
    board = []

    for row in state:
        board.append(row[:])

    x, y = find_blank(board)

    nx = x
    ny = y

    if action == "UP":
        nx -= 1
    elif action == "DOWN":
        nx += 1
    elif action == "LEFT":
        ny -= 1
    elif action == "RIGHT":
        ny += 1

    board[x][y], board[nx][ny] = board[nx][ny], board[x][y]
    return board
```

```python
def child_node(node, action):
    new_state = result(node.state, action)
    return Node(new_state, node, action)
```

```python
def solution(node):
    path = []

    while node.parent is not None:
        path.append(node.action)
        node = node.parent
    path.reverse()

    return path
```

```python
def breadth_first_search(initial_state):
    node = Node(initial_state)

    if goal_test(node.state):
        return solution(node)

    frontier = deque()

    frontier.append(node)

    reached = []

    while len(frontier) > 0:
        node = frontier.popleft()

        reached.append(node.state)

        if goal_test(node.state):
            return solution(node)

        for action in actions(node.state):
            child = child_node(node, action)

            in_reached = False

            for s in reached:
                if s == child.state:
                    in_reached = True

            in_frontier = False

            for n in frontier:
                if n.state == child.state:
                    in_frontier = True

            if not in_reached and not in_frontier:
                frontier.append(child)

    return None
```

```python
board = [
    [1,2,3],
    [4,5,6],
    [7,8,0]
]

# xáo trộn bằng các bước hợp lệ
for _ in range(20):
    possible = actions(board)
    action = random.choice(possible)
    board = result(board, action)

print("INITIAL STATE:")

for row in board:
    print(row)

print("-" * 30)

path = breadth_first_search(board)

if path is None:
    print("Không tìm thấy lời giải!")
else:
    print("SOLUTION:")

    for step, action in enumerate(path):
        print("Step", step + 1, ":", action)
```
