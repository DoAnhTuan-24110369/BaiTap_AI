State: 
0 = Sạch, 
1 = Bụi, 
2 = Robot, 
3 = Chướng ngại vật

```python
import random
```

```python
def rules(matrix, state):
    x = state[0]
    y = state[1]
    move = []

    if x > 0 and matrix[x-1][y] != 3:
        move.append("UP")

    if x < 2 and matrix[x+1][y] != 3:
        move.append("DOWN")

    if y > 0 and matrix[x][y-1] != 3:
        move.append("LEFT")

    if y < 2 and matrix[x][y+1] != 3:
        move.append("RIGHT")
    
    # Robot bị kẹt, không thể di chuyển
    if len(move) == 0:
        return "STAY"

    action = random.choice(move)

    return action

    
```

```python
def interpret_input(percept):
    state = []

    for i in range(3):
        for j in range(3):
            if percept[i][j] == 2:
                state = [i, j]

    return state
```

```python
def move_robot(matrix, state, action):
    if action == "STAY":
        return
    
    x = state[0]
    y = state[1]

    matrix[x][y] = 0

    if action == 'UP':
        x -= 1
    elif action == 'DOWN':
        x += 1
    elif action == 'LEFT':
        y -= 1
    elif action == 'RIGHT':
        y += 1

    matrix[x][y] = 2
```

```python
def operation(matrix):
    state = interpret_input(matrix)
    action = rules(matrix, state)
    move_robot(matrix, state, action)
    return action
```

```python
matrix = [[1, 1, 1],
        [1, 1, 1],
        [1, 1, 1]]

# Vị trí máy hút bụi ngẫu nhiên
matrix[random.randint(0,2)][random.randint(0,2)] = 2

# Vị trí chướng ngại vật ngẫu nhiên (2 cái)
for _ in range(2):
    if matrix[random.randint(0,2)][random.randint(0,2)] == 1:
        matrix[random.randint(0,2)][random.randint(0,2)] = 3

step = 0
while True:
    print("Step: ", step)
    print("Trạng thái:")
    for row in matrix:
        print(row)

    action = operation(matrix)

    print("-" * 30)
    print("Hành động:", action)
    print("\n")

    count = 0
    for row in matrix:
        count += row.count(1)

    if count == 0:
        print("Đã hút sạch bụi!")
        for row in matrix:
            print(row)
        break
    
    step += 1
```
