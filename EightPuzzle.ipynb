```python
import random
```

```python
def rules(state):
    move = []

    x = state[0]
    y = state[1]

    if x > 0:
        move.append("UP")
    if x < 2:
        move.append("DOWN")
    if y > 0:
        move.append("LEFT")
    if y < 2:
        move.append("RIGHT")

    return random.choice(move)
```

```python
def goal(board):
    win = [
        [1,2,3],
        [4,5,6],
        [7,8,0]
    ]

    return board == win
```

```python
def find_blank(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                return [i, j]
```

```python
def move(board, state, action):
    x = state[0]
    y = state[1]

    nx = x
    ny = y

    if action == "UP":
        nx -= 1
    elif action == "DOWN":
        nx += 1
    elif action == "LEFT":
        ny -= 1
    elif action == "RIGHT":
        ny += 1

    board[x][y], board[nx][ny] = board[nx][ny], board[x][y]
```

```python
def operation(board):
    state = find_blank(board)
    action = rules(state)
    move(board, state, action)
    return action
```

```python
nums = [1,2,3,4,5,6,7,8,0]
random.shuffle(nums)

board = []
k = 0
for i in range(3):
    row = []
    for j in range(3):
        row.append(nums[k])
        k += 1
    board.append(row)
    
step = 0
while step < 50:
    for row in board:
        print(row)
    print("-" * 25)

    if goal(board):
        print("Đã giải xong Puzzle!")
        break

    action = operation(board)
    print("Hành động:", action)
    
    step += 1
```
