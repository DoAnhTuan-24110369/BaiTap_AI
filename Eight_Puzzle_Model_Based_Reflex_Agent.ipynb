```python
import random
```

```python
def rules(state):
    move = []

    x = state[0]
    y = state[1]

    if x > 0:
        move.append("UP")
    if x < 2:
        move.append("DOWN")
    if y > 0:
        move.append("LEFT")
    if y < 2:
        move.append("RIGHT")

    return move
```

```python
def goal(board):
    win = [
        [1,2,3],
        [4,5,6],
        [7,8,0]
    ]

    return board == win
```

```python
def find_blank(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                return [i, j]
```

```python
def update_state(state, action, percept, model):
    new_state = find_blank(percept)
    
    model["forbidden"] = None
    
    if action == "UP":
        model["forbidden"] = "DOWN"

    elif action == "DOWN":
        model["forbidden"] = "UP"

    elif action == "LEFT":
        model["forbidden"] = "RIGHT"

    elif action == "RIGHT":
        model["forbidden"] = "LEFT"
    
    return new_state
```

```python
def rule_match(state, model):
    move = rules(state)
    forbidden = model["forbidden"]
    
    if forbidden in move:
        move.remove(forbidden)
    
    return random.choice(move)
```

```python
def move(board, state, action):
    x = state[0]
    y = state[1]

    nx = x
    ny = y

    if action == "UP":
        nx -= 1
    elif action == "DOWN":
        nx += 1
    elif action == "LEFT":
        ny -= 1
    elif action == "RIGHT":
        ny += 1

    board[x][y], board[nx][ny] = board[nx][ny], board[x][y]
```

```python
def operation(board, state, action, model):
    percept = board
    state = update_state(state, action, percept, model)
    action = rule_match(state, model)
    move(board, state, action)
    return state, action
```

```python
nums = [1,2,3,4,5,6,7,8,0]
random.shuffle(nums)

board = []
k = 0
for i in range(3):
    row = []
    for j in range(3):
        row.append(nums[k])
        k += 1
    board.append(row)
    
model = {
    "forbidden": None
}

state = find_blank(board)

action = None

step = 0
while step < 100:
    print("Step:", step)
    
    for row in board:
        print(row)
    print("-" * 25)

    if goal(board):
        print("Đã giải xong Puzzle!")
        break

    state, action = operation(board, state, action, model)
    print("Hành động:", action)
    print("\n")
    
    step += 1
```
