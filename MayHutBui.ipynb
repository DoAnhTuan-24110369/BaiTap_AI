```python
import random
```

```python
def rules(matrix, state):
    x = state[0]
    y = state[1]
    move = []

    if x > 0:
        move.append('UP')
    if x < 2:
        move.append('DOWN')
    if y > 0:
        move.append('LEFT')
    if y < 2:
        move.append('RIGHT')
        
    if matrix[x][y] == 1:
        matrix[x][y] = 0

    action = random.choice(move)

    return action

    
```

```python
def interpret_input(percept):
    state = []

    for i in range(3):
        for j in range(3):
            if percept[i][j] == 2:
                state = [i, j]

    return state
```

```python
def move_robot(matrix, state, action):
    x = state[0]
    y = state[1]

    matrix[x][y] = 0

    if action == 'UP':
        x -= 1
    elif action == 'DOWN':
        x += 1
    elif action == 'LEFT':
        y -= 1
    elif action == 'RIGHT':
        y += 1

    matrix[x][y] = 2
```

```python
def operation(matrix):
    state = interpret_input(matrix)
    action = rules(matrix, state)
    move_robot(matrix, state, action)
    return action
```

```python
matrix = [[1, 1, 1],
        [1, 1, 1],
        [1, 1, 1]]
matrix[random.randint(0,2)][random.randint(0,2)] = 2 #Vi tri may hut bui

while True:
    print("Trạng thái:")
    for row in matrix:
        print(row)

    action = operation(matrix)

    print("Máy hút bụi lựa chọn:", action)
    print("-" * 30)

    count = 0
    for row in matrix:
        count += row.count(1)

    if count == 0:
        print("Đã hút sạch bụi!")
        for row in matrix:
            print(row)
        break
```
